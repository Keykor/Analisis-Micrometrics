```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.preprocessing import StandardScaler

def evaluate(model, x, y):
    y_pred = model.predict(x)
    return {
        'MAE': metrics.mean_absolute_error(y_pred,y),
        'MSE': metrics.mean_squared_error(y_pred,y),
        'R^2': metrics.r2_score(y,y_pred)
    }
```

```python
feature_cols = ['avgFocusDuration', 'avgPauseDuration', 'avgSpeedPerTest', 'clicksPerTest',	'timePerTest', 'totalFocus', 'totalPauses', 'avgIntratecla', 'totalBorrado']
#feature_cols = ['timePerTest', 'totalFocus', 'avgIntratecla']
#feature_cols = ['avgFocusDuration', 'avgPauseDuration', 'avgSpeedPerTest', 'clicksPerTest',	'totalFocus', 'totalPauses', 'avgIntratecla', 'totalBorrado']
X_df = df_filtrado.loc[:, feature_cols]
X_df.shape
```

```python
y = df_filtrado['level']
y.shape
```

```python
X = X_df.to_numpy()
y = y.to_numpy()
```

### Entrenamiento "normal" ###

```python
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3)
fold_number = 1
list_r2 = []
list_mae = []
list_mse = []
for train, test in cv.split(X, y):
    x_train = X[train]
    y_train = y[train]
    x_test = X[test]
    y_test = y[test]

    sd = StandardScaler()
    sd.fit(x_train)
    x_train = sd.transform(x_train)
    x_test = sd.transform(x_test)

    model = LinearRegression()
    #model = DecisionTreeClassifier()
    model.fit(x_train, y_train)

    print("------------------------------------------")
    print("FOLD NUMBER " + str(fold_number))
    fold_number += 1

    training_result = evaluate(model, x_train, y_train)
    print("Training evaluate results")
    print(training_result)

    testing_result = evaluate(model, x_test, y_test)
    print("Testing evaluate results")
    print(testing_result)
    list_r2.append(testing_result['R^2'])
    list_mae.append(testing_result['MAE'])
    list_mse.append(testing_result['MSE'])

    print("------------------------------------------")
    print("")

print('R^2: %.3f (%.3f)' % (np.mean(list_r2), np.std(list_r2)))
print('MAE: %.3f (%.3f)' % (np.mean(list_mae), np.std(list_mae)))
print('MSE: %.3f (%.3f)' % (np.mean(list_mse), np.std(list_mse)))
```

### Entrenamiento con SelectFromModel ###

```python
from sklearn.feature_selection import SelectFromModel

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3)
fold_number = 1
list_r2 = []
list_mae = []
list_mse = []
for train, test in cv.split(X, y):
    x_train = X[train]
    y_train = y[train]
    x_test = X[test]
    y_test = y[test]

    sd = StandardScaler()
    sd.fit(x_train)
    x_train = sd.transform(x_train)
    x_test = sd.transform(x_test)

    model = SelectFromModel(LinearRegression())
    model.fit(x_train, y_train)

    print("------------------------------------------")
    print("FOLD NUMBER " + str(fold_number))
    fold_number += 1

    training_result = evaluate(model.estimator_, x_train, y_train)
    print("Training evaluate results")
    print(training_result)

    testing_result = evaluate(model.estimator_, x_test, y_test)
    print("Testing evaluate results")
    print(testing_result)
    list_r2.append(testing_result['R^2'])
    list_mae.append(testing_result['MAE'])
    list_mse.append(testing_result['MSE'])

    print("Columns")
    column_names = X_df.columns[model.get_support()]
    print(column_names)

    print("------------------------------------------")
    print("")

print('R^2: %.3f (%.3f)' % (np.mean(list_r2), np.std(list_r2)))
print('MAE: %.3f (%.3f)' % (np.mean(list_mae), np.std(list_mae)))
print('MSE: %.3f (%.3f)' % (np.mean(list_mse), np.std(list_mse)))
```
